```python
from selenium import webdriver
import time

# 看准网_百度
url = 'https://www.kanzhun.com/job/g11514.html?ka=paging1'
browser = webdriver.Chrome() 
browser.get(url)
time.sleep(0.5)
```

```python
def scrap_baidu_info():
    all_infomation = browser.find_element_by_class_name("f_l_con")
    corporations = all_infomation.find_elements_by_class_name("recruitment_item")
    all_corporation = []
    for corporation in corporations:
        corp_info = {}
        corp_info['name'] = corporation.find_element_by_class_name("name").text
        corp_info['salary'] = corporation.find_element_by_class_name("salary").text
        all_corporation.append(corp_info)
    return all_corporation
```

```python
all_info = []
for i in range(0,9):
    time.sleep(2) 
    all_info.extend(scrap_baidu_info())
    element = browser.find_element_by_class_name("next")
    time.sleep(1)
    loc = element.location
    browser.execute_script("window.scrollTo({x}, {y});".format(**loc))
    element.click()
    time.sleep(1)
all_info
```

```python
import pandas as pd
df1 = pd.DataFrame(all_info)
len(df1) # 178
```

```python
df1.to_csv("kanzhun_baidu.csv")
```
