```python
from selenium import webdriver
import time

# 看准网公司（互联网行业公司）
url = 'https://www.kanzhun.com/plc52p1.html?ka=select-hangye-1'
browser = webdriver.Chrome() 
browser.get(url)
time.sleep(0.5)
```

```python
def scrap_corporations_info():
    all_infomation = browser.find_element_by_class_name("company_result ")
    corporations = all_infomation.find_elements_by_tag_name("li")
    corporations
    all_corporation = []
    for corporation in corporations:
        corp_info = {}
        corp_info['name'] = corporation.find_elements_by_tag_name("a")[1].text
        corp_info['info'] = corporation.find_element_by_tag_name("div").find_element_by_tag_name("p").text#.split("|")
        corp_info['salary'] = corporation.find_element_by_tag_name("dl").find_element_by_class_name("grey_99").text
        corp_info['score'] = corporation.find_element_by_tag_name("dl").find_element_by_tag_name("dd").text
        all_corporation.append(corp_info)
    return all_corporation
# corp_info['info']
# a = ','.join(a)
```

```python
all_info = []
for i in range(0,5):
    time.sleep(1) 
    all_info.extend(scrap_corporations_info())
    element = browser.find_element_by_class_name("p_next")
    loc = element.location
    browser.execute_script("window.scrollTo({x}, {y});".format(**loc))
    time.sleep(1)
    element.click()
```

```python
import pandas as pd
df = pd.DataFrame(all_info)
df.to_csv("kanzhun_corp_internet.csv")
```
